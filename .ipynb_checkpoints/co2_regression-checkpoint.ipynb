### Imports

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

### Dataset

```python
co2 = pd.read_table('lamp_week.txt', header=None, sep=' ', names=['Dates', 'CO2'])
t = np.arange(1, len(co2['Dates']) * 7, 7)
year = np.sin((2 * np.pi / 365) * t)
semi = np.sin((2 * np.pi / 182) * t)
year_cos = np.cos((2 * np.pi / 365) * t)
semi_cos = np.cos((2 * np.pi / 182) * t)
t2 = t ** 2
co2['Time'] = pd.Series(t, index=co2.index)
co2['Time2'] = pd.Series(t2, index=co2.index)
co2['1 Year'] = pd.Series(year, index=co2.index)
co2['6 Months'] = pd.Series(semi, index=co2.index)
co2['1 Year Cos'] = pd.Series(year_cos, index=co2.index)
co2['6 Months Cos'] = pd.Series(semi_cos, index=co2.index)
co2 = co2[['Dates', 'Time', 'Time2', 'CO2', '1 Year', '6 Months', '1 Year Cos', '6 Months Cos']]
co2.head()
```

### Delete rows with NaNs

```python
co2NoNaN = co2.dropna()
co2NoNaN.head()
```

### Linear Regression

```python
### linear regression between time and co2
###
### simple
###
lin_reg = LinearRegression()
lin_reg.fit(co2NoNaN[['Time']], co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(lin_reg.intercept_, lin_reg.coef_))
plt.subplot(221)
plt.plot(co2NoNaN['Time'], co2NoNaN['CO2'], '-b')
plt.plot(co2NoNaN['Time'], lin_reg.intercept_ + lin_reg.coef_[0] * co2NoNaN['Time'], '-r', linewidth=2)
plt.xlim(co2NoNaN['Time'].iloc[0], co2NoNaN['Time'].iloc[-1])
###
### simple (regularized)
###
l2_lin_reg = Ridge(alpha=1, solver='cholesky')
l2_lin_reg.fit(co2NoNaN[['Time']], co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(l2_lin_reg.intercept_, l2_lin_reg.coef_))
plt.subplot(222)
plt.plot(co2NoNaN['Time'], co2NoNaN['CO2'], '-b')
plt.plot(co2NoNaN['Time'], l2_lin_reg.intercept_ + l2_lin_reg.coef_[0] * co2NoNaN['Time'], '-r', linewidth=2)
plt.xlim(co2NoNaN['Time'].iloc[0], co2NoNaN['Time'].iloc[-1])
###
### gradient descent
###
scaler = StandardScaler().fit(co2NoNaN[['Time']])
timeX = scaler.transform(co2NoNaN[['Time']])
sgd_reg = SGDRegressor(max_iter=50, tol=0.001, penalty=None, eta0=0.1)
sgd_reg.fit(timeX, co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(sgd_reg.intercept_, sgd_reg.coef_))
#T = scaler.inverse_transform(timeX)
plt.subplot(223)
plt.plot(timeX, co2NoNaN['CO2'], '-b')
plt.plot(timeX, sgd_reg.intercept_ + sgd_reg.coef_[0] * timeX, '-r', linewidth=2)
plt.xlim(timeX[0], timeX[-1])
###
### gradient descent (regularized)
###
scaler = StandardScaler().fit(co2NoNaN[['Time']])
timeX = scaler.transform(co2NoNaN[['Time']])
l2_sgd_reg = SGDRegressor(max_iter=50, tol=0.001, penalty='l2', eta0=0.1)
l2_sgd_reg.fit(timeX, co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(l2_sgd_reg.intercept_, l2_sgd_reg.coef_))
#T = scaler.inverse_transform(timeX)
plt.subplot(224)
plt.plot(timeX, co2NoNaN['CO2'], '-b')
plt.plot(timeX, l2_sgd_reg.intercept_ + l2_sgd_reg.coef_[0] * timeX, '-r', linewidth=2)
plt.xlim(timeX[0], timeX[-1])
```

```python
### linear regression between (time, 1 year, 6 months) and co2
###
### simple
###
lin_reg = LinearRegression()
lin_reg.fit(co2NoNaN[['Time', '1 Year', '6 Months']], co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(lin_reg.intercept_, lin_reg.coef_))
plt.subplot(221)
plt.plot(co2NoNaN['Time'], co2NoNaN['CO2'], '-b')
plt.plot(co2NoNaN['Time'], lin_reg.intercept_ + lin_reg.coef_[0] * co2NoNaN['Time'] +
         lin_reg.coef_[1] * co2NoNaN['1 Year'] + lin_reg.coef_[2] * co2NoNaN['6 Months'], '-r', linewidth=2)
plt.xlim(co2NoNaN['Time'].iloc[0], co2NoNaN['Time'].iloc[-1])
###
### simple (regularized)
###
l2_lin_reg = Ridge(alpha=1, solver='cholesky')
l2_lin_reg.fit(co2NoNaN[['Time', '1 Year', '6 Months']], co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(l2_lin_reg.intercept_, l2_lin_reg.coef_))
plt.subplot(222)
plt.plot(co2NoNaN['Time'], co2NoNaN['CO2'], '-b')
plt.plot(co2NoNaN['Time'], l2_lin_reg.intercept_ + l2_lin_reg.coef_[0] * co2NoNaN['Time'] +
         l2_lin_reg.coef_[1] * co2NoNaN['1 Year'] + l2_lin_reg.coef_[2] * co2NoNaN['6 Months'], '-r', linewidth=2)
plt.xlim(co2NoNaN['Time'].iloc[0], co2NoNaN['Time'].iloc[-1])
###
### gradient descent
###
X = co2NoNaN[['Time', '1 Year', '6 Months']]
scaler = StandardScaler().fit(X)
newX = scaler.transform(X)
sgd_reg = SGDRegressor(max_iter=50, tol=0.001, penalty=None, eta0=0.1)
sgd_reg.fit(newX, co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(sgd_reg.intercept_, sgd_reg.coef_))
plt.subplot(223)
plt.plot(newX[:,0], co2NoNaN['CO2'], '-b')
plt.plot(newX[:,0], sgd_reg.intercept_ + sgd_reg.coef_[0] * newX[:,0] +
         sgd_reg.coef_[1] * newX[:,1] + sgd_reg.coef_[2] * newX[:,2], '-r', linewidth=2)
plt.xlim(newX[0,0], newX[-1,0])
###
### gradient descent (regularized)
###
X = co2NoNaN[['Time', '1 Year', '6 Months']]
scaler = StandardScaler().fit(X)
newX = scaler.transform(X)
l2_sgd_reg = SGDRegressor(max_iter=50, tol=0.001, penalty=None, eta0=0.1)
l2_sgd_reg.fit(newX, co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(l2_sgd_reg.intercept_, l2_sgd_reg.coef_))
plt.subplot(224)
plt.plot(newX[:,0], co2NoNaN['CO2'], '-b')
plt.plot(newX[:,0], l2_sgd_reg.intercept_ + l2_sgd_reg.coef_[0] * newX[:,0] +
         l2_sgd_reg.coef_[1] * newX[:,1] + l2_sgd_reg.coef_[2] * newX[:,2], '-r', linewidth=2)
plt.xlim(newX[0,0], newX[-1,0])
```

### Polynomial Regression

```python
### polynomial regression
###
### simple
###
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(co2NoNaN[['Time']])
lin_reg = LinearRegression()
lin_reg.fit(X_poly, co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(lin_reg.intercept_, lin_reg.coef_))
plt.subplot(221)
plt.plot(X_poly[:,0], co2NoNaN['CO2'], '-b')
plt.plot(X_poly[:,0], lin_reg.intercept_ + X_poly.dot(lin_reg.coef_), '-r', linewidth=2)
plt.xlim(X_poly[0,0], X_poly[-1,0])
###
### simple (regularized)
###
poly_features = PolynomialFeatures(degree=5, include_bias=False)
X_poly = poly_features.fit_transform(co2NoNaN[['Time']])
lin_reg = Ridge(alpha=1, solver='cholesky')
lin_reg.fit(X_poly, co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(lin_reg.intercept_, lin_reg.coef_))
plt.subplot(222)
plt.plot(X_poly[:,0], co2NoNaN['CO2'], '-b')
plt.plot(X_poly[:,0], lin_reg.intercept_ + X_poly.dot(lin_reg.coef_), '-r', linewidth=2)
plt.xlim(X_poly[0,0], X_poly[-1,0])
###
### polynomial with sinusoids (experiment)
###
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(co2NoNaN[['Time', '1 Year', '6 Months']])
lin_reg = LinearRegression()
lin_reg.fit(X_poly, co2NoNaN['CO2'])
print('Best fitting line: {} + {} * t'.format(lin_reg.intercept_, lin_reg.coef_))
plt.subplot(223)
plt.plot(X_poly[:,0], co2NoNaN['CO2'], '-b')
plt.plot(X_poly[:,0], lin_reg.intercept_ + X_poly.dot(lin_reg.coef_), '-r', linewidth=2)
plt.xlim(X_poly[0,0], X_poly[-1,0])
```

### Non Linear Regression

```python
### non linear regression
###
### simple
###
X = co2NoNaN[['Time', 'Time2', '1 Year', '6 Months', '1 Year Cos', '6 Months Cos']]
lin_reg = LinearRegression()
lin_reg.fit(X, co2NoNaN['CO2'])
r2 = r2_score(co2NoNaN['CO2'], lin_reg.predict(X))
print('Best fitting line: {} + {} * t'.format(lin_reg.intercept_, lin_reg.coef_))
plt.subplot(221)
plt.plot(co2NoNaN['Time'], co2NoNaN['CO2'], '-b')
plt.plot(co2NoNaN['Time'], lin_reg.intercept_ + X.dot(lin_reg.coef_), '-r', linewidth=2)
plt.xlim(co2NoNaN['Time'].iloc[0], co2NoNaN['Time'].iloc[-1])
plt.title('R2 = {:.8f}'.format(r2))
###
### regularized
###
X = co2NoNaN[['Time', 'Time2', '1 Year', '6 Months', '1 Year Cos', '6 Months Cos']]
lin_reg = Ridge(alpha=1, solver='cholesky')
lin_reg.fit(X, co2NoNaN['CO2'])
r2 = r2_score(co2NoNaN['CO2'], lin_reg.predict(X))
print('Best fitting line: {} + {} * t'.format(lin_reg.intercept_, lin_reg.coef_))
plt.subplot(222)
plt.plot(co2NoNaN['Time'], co2NoNaN['CO2'], '-b')
plt.plot(co2NoNaN['Time'], lin_reg.intercept_ + X.dot(lin_reg.coef_), '-r', linewidth=2)
plt.xlim(co2NoNaN['Time'].iloc[0], co2NoNaN['Time'].iloc[-1])
plt.title('R2 = {:.8f}'.format(r2))
```

### Outliers

```python
from sklearn.svm import LinearSVR
from sklearn.neighbors import LocalOutlierFactor
from sklearn.covariance import EllipticEnvelope
from sklearn.svm import OneClassSVM
```

```python
X = co2NoNaN[['Time', 'Time2', '1 Year', '6 Months', '1 Year Cos', '6 Months Cos']]
scaler = StandardScaler().fit(X)
newX = scaler.transform(X)
out_lim = 10
svm_reg = LinearSVR(epsilon=out_lim)
svm_reg.fit(newX, co2NoNaN['CO2'])
r2 = r2_score(co2NoNaN['CO2'], svm_reg.predict(newX))
fit = svm_reg.intercept_ + newX.dot(svm_reg.coef_) + out_lim
fit_low = svm_reg.intercept_ + newX.dot(svm_reg.coef_)
fit_high = svm_reg.intercept_ + newX.dot(svm_reg.coef_) + 2*out_lim
fig, ax = plt.subplots()
plt.plot(X['Time'], co2NoNaN['CO2'], '-b')
plt.plot(X['Time'], fit, '-r', linewidth=2)
ax.fill_between(X['Time'], fit_high, fit_low, color="crimson", alpha=0.3)
plt.xlim(X['Time'].iloc[0], X['Time'].iloc[-1])
plt.title('R2 = {:.8f}'.format(r2))
```

```python
res = pd.read_table('res.txt', header=None, sep=' ', names=['Time', 'Res', 'Res Norm', 'Dates'])
res.head()
```

```python
resNoNaN = res.dropna()
resNoNaN.head()
```

```python
X = resNoNaN[['Time']]
scaler = StandardScaler().fit(X)
newX = scaler.transform(X)
out_lim = 3
line = np.zeros((len(newX)), dtype=float)
svm_reg = LinearSVR(epsilon=out_lim)
svm_reg.fit(newX, line)
fit = svm_reg.intercept_ + newX.dot(svm_reg.coef_)
fit_low = svm_reg.intercept_ + newX.dot(svm_reg.coef_) - out_lim
fit_high = svm_reg.intercept_ + newX.dot(svm_reg.coef_) + out_lim
res_scatter = resNoNaN[(resNoNaN['Res Norm'] > 3.0) | (resNoNaN['Res Norm'] < -3.0)]
fig, ax = plt.subplots()
plt.plot(X['Time'], resNoNaN['Res Norm'], '-b', label='CO2')
plt.scatter(res_scatter['Time'], res_scatter['Res Norm'], s=100, edgecolors='k',
            facecolors='none', label='Outlier scores')
ax.fill_between(X['Time'], fit_high, fit_low, color="crimson", alpha=0.3)
plt.xlim(X['Time'].iloc[0], X['Time'].iloc[-1])
```

```python
X = co2NoNaN[['CO2']]
clf = LocalOutlierFactor(n_neighbors=100, contamination=0.1)
y_pred = clf.fit_predict(X)
X_scores = clf.negative_outlier_factor_
plt.title("Local Outlier Factor (LOF)")
plt.scatter(co2NoNaN['Time'], co2NoNaN['CO2'], color='k', s=3., label='CO2')
# plot circles with radius proportional to the outlier scores
radius = (X_scores.max() - X_scores) / (X_scores.max() - X_scores.min())
plt.scatter(co2NoNaN['Time'], co2NoNaN['CO2'], s=300 * radius, edgecolors='r',
            facecolors='none', label='Outlier scores')
plt.legend()
```

```python
X = resNoNaN[['Res Norm']]
clf = LocalOutlierFactor(n_neighbors=100, contamination=0.1)
y_pred = clf.fit_predict(X)
X_scores = clf.negative_outlier_factor_
plt.title("Local Outlier Factor (LOF)")
plt.scatter(resNoNaN['Time'], resNoNaN['Res Norm'], color='k', s=3., label='CO2')
# plot circles with radius proportional to the outlier scores
radius = (X_scores.max() - X_scores) / (X_scores.max() - X_scores.min())
size = 300 * radius
for i in range(len(resNoNaN['Res Norm'])):
    if size[i] > 50.0:
        plt.scatter(resNoNaN['Time'].iloc[i], resNoNaN['Res Norm'].iloc[i], s=size[i], edgecolors='r',
                    facecolors='none')#, label='Outlier scores')
plt.ylim(-4, 6)
plt.legend()
```

```python
X = resNoNaN[['Time', 'Res Norm']]
clf = EllipticEnvelope(contamination=0.261)
xx, yy = np.meshgrid(np.linspace(-500, 9000, 500), np.linspace(-4, 6, 500))
clf.fit(X)
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.subplot(211)
plt.contour(xx, yy, Z, levels=[0], linewidths=3, colors='r')
plt.scatter(resNoNaN['Time'], resNoNaN['Res Norm'], s=3., color='black')
plt.title('Res Norm')
X2 = resNoNaN[['Time', 'Res']]
clf2 = EllipticEnvelope(contamination=0.261)
xx2, yy2 = np.meshgrid(np.linspace(-500, 9000, 500), np.linspace(-4, 6, 500))
clf2.fit(X2)
Z2 = clf2.decision_function(np.c_[xx2.ravel(), yy2.ravel()])
Z2 = Z2.reshape(xx2.shape)
plt.subplot(212)
plt.contour(xx2, yy2, Z2, levels=[0], linewidths=3, colors='r')
plt.scatter(resNoNaN['Time'], resNoNaN['Res'], s=3., color='black')
plt.title('Res')
plt.subplots_adjust(hspace=0.5)
```

```python
X = resNoNaN[['Time', 'Res Norm']]
clf = OneClassSVM(nu=0.1, kernel="rbf", gamma=0.1)
xx, yy = np.meshgrid(np.linspace(-500, 9000, 500), np.linspace(-4, 6, 500))
clf.fit(X)
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.subplot(211)
plt.contour(xx, yy, Z, levels=[0], linewidths=3, colors='r')
plt.scatter(resNoNaN['Time'], resNoNaN['Res Norm'], s=3., color='black')
plt.title('Res Norm')
X2 = resNoNaN[['Time', 'Res']]
clf2 = OneClassSVM(nu=0.461, gamma=0.05)
xx2, yy2 = np.meshgrid(np.linspace(-500, 9000, 500), np.linspace(-4, 6, 500))
clf2.fit(X2)
Z2 = clf2.decision_function(np.c_[xx2.ravel(), yy2.ravel()])
Z2 = Z2.reshape(xx2.shape)
plt.subplot(212)
plt.contour(xx2, yy2, Z2, levels=[0], linewidths=3, colors='r')
plt.scatter(resNoNaN['Time'], resNoNaN['Res'], s=3., color='black')
plt.title('Res')
plt.subplots_adjust(hspace=0.5)
```
