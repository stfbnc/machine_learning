## Imports

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="white")
%matplotlib inline
```

# White - Whink - Pink - Pired - Red noise with clustering

## Loading data

```python
wpr = pd.read_table('./noise/clf_noise5.txt', header=None, sep=' ', names=['Hurst', 'Length', 'Colour_idx', 'Colour'])
wpr.head()
```

```python
L_plot = wpr['Length']
H_plot = wpr['Hurst']
plt.plot(L_plot[wpr['Colour_idx'] == 0], H_plot[wpr['Colour_idx'] == 0], 'ok', label='$H_{wn}$')
plt.plot(L_plot[wpr['Colour_idx'] == 1], H_plot[wpr['Colour_idx'] == 1], 'oc', label='$H_{wpn}$')
plt.plot(L_plot[wpr['Colour_idx'] == 2], H_plot[wpr['Colour_idx'] == 2], 'om', label='$H_{pn}$')
plt.plot(L_plot[wpr['Colour_idx'] == 3], H_plot[wpr['Colour_idx'] == 3], 'og', label='$H_{prn}$')
plt.plot(L_plot[wpr['Colour_idx'] == 4], H_plot[wpr['Colour_idx'] == 4], 'or', label='$H_{rn}$')
plt.xlim(-200, 10300)
plt.ylim(0.3, 1.8)
plt.title('Hurst exponent for white, whink, pink, pired, and red noises')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.8))
```

## Sklearn imports

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture
```

## Data

```python
X = wpr[['Length', 'Hurst']]
y = wpr['Colour_idx']
print('X shape: {}'.format(X.shape))
print('y shape: {}'.format(y.shape))
```

# K-means

## Feature scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Training

```python
init_cntrs = scaler.transform(np.array([[5000, 0.5], [5000, 0.75], [5000, 1.0], [5000, 1.25], [5000, 1.5]]))
kmeans = KMeans(n_clusters=5, random_state=0, init=init_cntrs)
kmeans.fit(X_scaled)
```

```python
### values of length and hurst
values = scaler.transform([[8004, 1.67]])
### predict for input values (returns 0 for white, 1 for whink, 2 for pink, 3 for pired, and 4 for red)
print('Predicted class: {}'.format('white' if kmeans.predict(values) == 0 else('whink' if kmeans.predict(values) == 1 else('pink' if kmeans.predict(values) == 2 else('pired' if kmeans.predict(values) == 3 else 'red')))))
```

## Decision Plot

```python
### decision boundary plot
lx = -200
rx = 10300
by = 0.3
ty = 1.8
step = 0.01
lim_new = scaler.transform([[lx, by], [rx, ty]])
xx, yy = np.mgrid[lim_new[0, 0]:lim_new[1, 0]+step:step, lim_new[0, 1]:lim_new[1, 1]+step:step]
pnts = np.c_[xx.ravel(), yy.ravel()]
probs = kmeans.predict(pnts).reshape(xx.shape)
mean_tr = scaler.mean_
std_tr = np.sqrt(scaler.var_)
xx_new = xx * std_tr[0] + mean_tr[0]
yy_new = yy * std_tr[1] + mean_tr[1]
# Create color maps
cmap_light = ListedColormap(['#FFFFFF', '#AAAAFF', '#FFAAFF', '#AAFFAA', '#FFAAAA'])
cmap_bold = ListedColormap(['#FFFFFF', '#0000FF', '#FF00FF', '#00FF00', '#FF0000'])
plt.figure(figsize=(10, 5))
plt.pcolormesh(xx_new, yy_new, probs, cmap=cmap_light)
plt.scatter(X['Length'], X['Hurst'], c=y, cmap=cmap_bold, edgecolor='k', s=25, linewidth=1)
cbar = plt.colorbar()
cbar.set_ticks([0.5, 1.35, 2.1, 2.95, 3.7])
cbar.ax.set_yticklabels(['white', 'whink', 'pink', 'pired', 'red'], rotation=90)
plt.xlim(lx, rx)
plt.ylim(by, ty)
plt.title('Hurst exponent for white, whink, pink, pired, and red noises (K-means)')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
```

## Training

```python
agg_cl = AgglomerativeClustering(n_clusters=5)
agg_cl.fit(X_scaled)
```

## Training

```python
dbscan = DBSCAN(eps=2, min_samples=2)
dbscan.fit(X_scaled)
```

# White - Pink - Red noise with clustering

## Loading data

```python
wpr = pd.read_table('./noise/clf_noise.txt', header=None, sep=' ', names=['Hurst', 'Length', 'Colour_idx', 'Colour'])
wpr.head()
```

```python
L_plot = wpr['Length']
H_plot = wpr['Hurst']
plt.plot(L_plot[wpr['Colour_idx'] == 0], H_plot[wpr['Colour_idx'] == 0], 'ok', label='$H_{wn}$')
plt.plot(L_plot[wpr['Colour_idx'] == 1], H_plot[wpr['Colour_idx'] == 1], 'om', label='$H_{pn}$')
plt.plot(L_plot[wpr['Colour_idx'] == 2], H_plot[wpr['Colour_idx'] == 2], 'or', label='$H_{rn}$')
plt.xlim(0, 3100)
plt.ylim(0.3, 1.8)
plt.title('Hurst exponent for white, pink, and red noises')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.9))
```

## Data

```python
X = wpr[['Length', 'Hurst']]
y = wpr['Colour_idx']
print('X shape: {}'.format(X.shape))
print('y shape: {}'.format(y.shape))
```

# K-means

## Feature scaling

```python
scaler = StandardScaler()
#idxs = np.arange(0, len(y), 10)
X_new = X#.iloc[idxs]
y_new = y#.iloc[idxs]
X_scaled = scaler.fit_transform(X_new)
```

## Training

```python
cntrs_x = 1000
init_cntrs = scaler.transform(np.array([[cntrs_x, 0.5], [cntrs_x, 1.0], [cntrs_x, 1.5]]))
kmeans3 = KMeans(n_clusters=3, random_state=0, init=init_cntrs)
kmeans3.fit(X_scaled)
```

```python
### values of length and hurst
values = scaler.transform([[1004, 1.27]])
### predict for input values (returns 0 for white, 1 for whink, 2 for pink, 3 for pired, and 4 for red)
print('Predicted class: {}'.format('white' if kmeans3.predict(values) == 0 else('pink' if kmeans3.predict(values) == 1 else 'red'))) 
```

## Decision Plot

```python
### decision boundary plot
lx = 0
rx = 3100
by = 0.3
ty = 1.8
step = 0.01
lim_new = scaler.transform([[lx, by], [rx, ty]])
xx, yy = np.mgrid[lim_new[0, 0]:lim_new[1, 0]+step:step, lim_new[0, 1]:lim_new[1, 1]+step:step]
pnts = np.c_[xx.ravel(), yy.ravel()]
probs = kmeans3.predict(pnts).reshape(xx.shape)
mean_tr = scaler.mean_
std_tr = np.sqrt(scaler.var_)
xx_new = xx * std_tr[0] + mean_tr[0]
yy_new = yy * std_tr[1] + mean_tr[1]
centers_x = kmeans3.cluster_centers_[:,0] * std_tr[0] + mean_tr[0]
centers_y = kmeans3.cluster_centers_[:,1] * std_tr[1] + mean_tr[1]
# Create color maps
cmap_light = ListedColormap(['#FFFFFF', '#FFAAFF', '#FFAAAA'])
cmap_bold = ListedColormap(['#FFFFFF', '#FF00FF', '#FF0000'])
plt.figure(figsize=(10, 5))
plt.pcolormesh(xx_new, yy_new, probs, cmap=cmap_light)
plt.scatter(X_new['Length'], X_new['Hurst'], c=y_new, cmap=cmap_bold, edgecolor='k', s=25, linewidth=1)
cbar = plt.colorbar()
cbar.set_ticks([0.33, 1.0, 1.66])
cbar.ax.set_yticklabels(['white', 'pink', 'red'], rotation=90)
plt.scatter(centers_x, centers_y, color='#00FF00', edgecolor='k', s=200, linewidth=1)
plt.xlim(lx, rx)
plt.ylim(by, ty)
plt.title('Hurst exponent for white, pink, and red noises (K-means)')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
```
