## Imports

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="white")
%matplotlib inline
```

# White - Pink noise classification

## Loading data

```python
wp = pd.read_table('./noise/clf_noise_nored.txt', header=None, sep=' ', names=['Hurst', 'Length', 'Colour_idx', 'Colour'])
wp.head()
```

```python
L_plot = wp['Length']
H_plot = wp['Hurst']
plt.plot(L_plot[wp['Colour_idx'] == 0], H_plot[wp['Colour_idx'] == 0], 'ok', label='$H_{wn}$')
plt.plot(L_plot[wp['Colour_idx'] == 1], H_plot[wp['Colour_idx'] == 1], 'om', label='$H_{pn}$')
plt.xlim(0, 3100)
plt.ylim(0.3, 1.3)
plt.title('Hurst exponent for white and pink noises')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.9))
```

## Sklearn imports

```python
from sklearn.model_selection import train_test_split, GridSearchCV, validation_curve, learning_curve
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, precision_score, precision_recall_curve, roc_curve, auc, classification_report, confusion_matrix, mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier, NearestNeighbors
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.decomposition import PCA, KernelPCA
```

## Train and test sets

```python
X = wp[['Length', 'Hurst']]
y = wp['Colour_idx']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print('X_train shape: {}'.format(X_train.shape))
print('y_train shape: {}'.format(y_train.shape))
print('X_test shape: {}'.format(X_test.shape))
print('y_test shape: {}'.format(y_test.shape))
```

## Train and test sets plots

```python
L_train = X_train['Length']
H_train = X_train['Hurst']
L_test = X_test['Length']
H_test = X_test['Hurst']
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(L_train[y_train == 0], H_train[y_train == 0], 'ok', label='$H_{wn}$ train')
plt.plot(L_train[y_train == 1], H_train[y_train == 1], 'om', label='$H_{pn}$ train')
plt.xlim(0, 3100)
plt.ylim(0.3, 1.3)
plt.title('Hurst exponent for white and pink noises (Train)')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
plt.legend(ncol=2)
plt.subplot(122)
plt.plot(L_test[y_test == 0], H_test[y_test == 0], 'ok', label='$H_{wn}$ test')
plt.plot(L_test[y_test == 1], H_test[y_test == 1], 'om', label='$H_{pn}$ test')
plt.xlim(0, 3100)
plt.ylim(0.3, 1.3)
plt.title('Hurst exponent for white and pink noises (Test)')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
plt.legend(ncol=2)
```

# Logistic regression

```python
### penalty = l1
log_reg = LogisticRegression()
grid_values = {'penalty': ['l1'], 'C':[0.01, 0.1, 1, 10, 100], 'solver':['liblinear']}
grid = GridSearchCV(log_reg, param_grid=grid_values, scoring='roc_auc', cv=3)
grid.fit(X_train, y_train)
print('Grid best parameter (AUC): {}'.format(grid.best_params_))
print('Grid best score (AUC): {}'.format(grid.best_score_))
grid = grid.best_estimator_
y_pred = grid.predict(X_test)
acc = grid.score(X_test, y_test)
rec = recall_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
conf_mtx = confusion_matrix(y_test, y_pred)
print('Accuracy: {}'.format(acc))
print('Recall: {}'.format(rec))
print('Precision: {}'.format(prec))
print(classification_report(y_test, y_pred))
print(conf_mtx)
```

```python
### ROC curve
precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
fpr, tpr, _ = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.xlim([0.0, 1.01])
plt.ylim([0.0, 1.01])
plt.plot(precision, recall, label='Precision-Recall Curve')
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.subplot(122)
plt.xlim([-0.01, 1.00])
plt.ylim([-0.01, 1.01])
plt.plot(fpr, tpr, lw=3, label='LogRegr ROC curve (area = {:0.2f})'.format(roc_auc))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], color='navy', lw=3, linestyle='--')
```

```python
### decision boundary plot
lx = 0
rx = 3100
by = 0.3
ty = 1.3
step = 0.01
xx, yy = np.mgrid[lx:rx+step:step, by:ty+step:step]
pnts = np.c_[xx.ravel(), yy.ravel()]
probs = grid.predict_proba(pnts)[:, 1].reshape(xx.shape)
f, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(xx, yy, probs, 50, cmap="RdBu", vmin=0, vmax=1)
ax_c = f.colorbar(contour)
ax_c.set_label("$P(noise = pink)$")
ax_c.set_ticks([0, 0.25, 0.5, 0.75, 1])
#ax.scatter(X['Length'], X['Hurst'], c=y, s=50, cmap="RdBu", vmin=-0.2, vmax=1.2, edgecolor="white", linewidth=1)
ax.scatter(X['Length']/y, X['Hurst']/y, c='m', s=50, edgecolor="white", linewidth=1)
ax.scatter(-X['Length']/(y-1), -X['Hurst']/(y-1), c='k', s=50, edgecolor="white", linewidth=1)
plt.xlim(lx, rx)
plt.ylim(by, ty)
plt.title('Hurst exponent for white and pink noises (Logistic Regression)')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
```

## SVM (correct procedure)

```python
# select the preprocessing pipeline operations and transform data
svm_pipe = Pipeline([('scaler', StandardScaler())])
X_new = svm_pipe.fit_transform(X_train)
# select and fit classifier
# it could be a GridSearchCV and then select the best one with clf.best_estimator_
clf = SVC(C=1, kernel='sigmoid')
clf.fit(X_new, y_train)
# transform and predict new values
X_test_new = svm_pipe.transform(X_test)
y_pred = clf.predict(X_test_new)
# metrics
print('MSE: {}'.format(np.sqrt(mean_squared_error(y_test, y_pred))))
print('Accuracy: {}'.format(clf.score(X_test, y_test)))
print('Recall: {}'.format(recall_score(y_test, y_pred)))
print('Precision: {}'.format(precision_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
### decision boundary plot
lx = 0
rx = 3100
by = 0.3
ty = 1.3
### transform limits for decision_function computation
lim_new = svm_pipe.transform([[lx, by], [rx, ty]])
ax = plt.gca()
### create grid to evaluate model
xx = np.linspace(lim_new[0, 0], lim_new[1, 0], 100)
yy = np.linspace(lim_new[0, 1], lim_new[1, 1], 100)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)
mean_tr = svm_pipe.named_steps['scaler'].mean_
std_tr = np.sqrt(svm_pipe.named_steps['scaler'].var_)
XX_new = XX * std_tr[0] + mean_tr[0]
YY_new = YY * std_tr[1] + mean_tr[1]
### no need to transform back Z, it corresponds to the un-transformed y
### plot decision boundary and margins
ax.contour(XX_new, YY_new, Z, colors='r', levels=[-1, 0, 1], linestyles=['--', '-', '--'], linewidths=3)
ax.scatter(X['Length']/y, X['Hurst']/y, c='m', s=50, edgecolor="white", linewidth=1)
ax.scatter(-X['Length']/(y-1), -X['Hurst']/(y-1), c='k', s=50, edgecolor="white", linewidth=1)
plt.xlim(lx, rx)
plt.ylim(by, ty)
plt.title('Hurst exponent for white and pink noises (SVM)')
plt.xlabel('Time series length')
plt.ylabel('Hurst exponent')
```

## Validation curve (SVM)

```python
### validation_curve is slow, so we take a subset of X and y
sz = 20
idxs = np.random.randint(y.shape[0], size=sz)
X_val = np.zeros((sz, 2))
y_val = np.zeros((sz, ))
for i in range(sz):
    X_val[i,0] = X['Length'].iloc[idxs[i]]
    X_val[i,1] = X['Hurst'].iloc[idxs[i]]
    y_val[i] = y[idxs[i]]
param_range = np.arange(0.01, 100, 20)
train_scores, test_scores = validation_curve(SVC(C=0.01, kernel='linear'), X_val, y_val,
                                                 param_name='C', param_range=param_range, scoring='precision')
```

```python
### plot validation_curve
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
plt.plot(param_range, train_scores_mean, label='Training score', color='darkorange')
plt.fill_between(param_range, train_scores_mean-train_scores_std, train_scores_mean+train_scores_std, alpha=0.2, color='darkorange')
plt.plot(param_range, test_scores_mean, label='Cross-validation score', color='navy')
plt.fill_between(param_range, test_scores_mean-test_scores_std, test_scores_mean+test_scores_std, alpha=0.2, color='navy')
plt.legend(loc=4)
plt.xlabel('C')
plt.ylabel('Precision')
```

```python
### learning curve
train_sizes, train_scores, test_scores = learning_curve(SVC(kernel='linear'), X_test, y_test, train_sizes=[50, 80, 100], cv=5, n_jobs=-1)
```

```python
### learning curve plot
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
plt.grid()
plt.fill_between(train_sizes, train_scores_mean-train_scores_std, train_scores_mean+train_scores_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, test_scores_mean-test_scores_std, test_scores_mean+test_scores_std, alpha=0.1, color='g')
plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Training score')
plt.plot(train_sizes, test_scores_mean, 'o-', color='g', label='Cross-validation score')
plt.legend(loc='best')
plt.xlabel('Training examples')
plt.ylabel('Score')
```
